```python
import cv2
from ultralytics import YOLO
import time
```

```python
model = YOLO('yolo11n.pt')
prev_time = 0
```

```python
imag = cv2.imread('C:/Users/nisha/OneDrive - Indian Institute of Technology Guwahati/Documents/projects/CCTV/data/crowd.jpg')
if imag is None:
    raise FileNotFoundError("Image not found. Please check the file path.")

lk = 0
res = model(imag)
for r in res:
    for box in r.boxes:
        lk +=1
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
        cv2.rectangle(imag, (x1, y1), (x2, y2), (0, 255, 231), 2)
        

while True:
    cv2.imshow('YOLO', imag)
    cv2.putText(imag,str(lk),(15,60),cv2.FONT_HERSHEY_PLAIN,4,(0,150,255),4)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()

        
        

```

```python
cap = cv2.VideoCapture(0)
while True:

    cont, frame = cap.read()
    
    if not cont:
        print('video samapt')
        break
    current_time = time.time()
    fps = 1 / (current_time - prev_time)
    prev_time = current_time

    results = model(frame)
    lk = 0
    for r in results:
        class_names = r.names
        for box in r.boxes:
            lk +=1
            x1,y1,x2,y2 = map(int,box.xyxy[0].tolist())
            confidence = float(box.conf[0])
            class1 = int(box.cls[0])
            label = box.cls[0]
            detected_class = class_names[class1]
            if detected_class.lower() == 'person':
                label = f"{confidence:.2f}"
                cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)
                cv2.putText(frame,str(confidence),(x1-10,y1-10),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,255,255),1)
                cv2.putText(frame,str(lk),(x1-15,y1-30),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,255,255),1)
                cv2.putText(frame,str(fps),(20,30),cv2.FONT_HERSHEY_PLAIN,2,(0,190,200),3)
            else:
                pass
    
    cv2.imshow('detected image',frame)

    if cv2.waitKey(1) & 0xFF == ord('q') :
        break

cap.release()
cv2.destroyAllWindows()
        
    
    
```
